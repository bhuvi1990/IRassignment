```
def preprocess_data(contents):
  datadict = {}
  uwords = {}
  #print(contents)
  for content in contents:
    sen = content[1]
    #print(sen)
    words = sen.split()
    #print(words)
    for word in words:
      if word not in uwords:
        uwords[word] = []
        uwords[word].append(1)
        uwords[word].append([content[0]])
      else:
        uwords[word][0]+= 1
        uwords[word][1].append(content[0])
      

    datadict[content[0]] = words

    #print(datadict)   
  return  uwords
    

```

```
def low(z):
  return z.lower()

def first(z):
  return z[0]

def squish(x):
  s = set(x)
  return sorted([(i,x.count(i)) for i in s], key = first)
```
